```python
#bibliotecas que serão utlizadas:
import pandas as pd #biblioteca de manipulção de dataframes
import numpy as np #manipulção de arrays
import matplotlib.pyplot as plt #graficos
import seaborn as sns #graficos
```

```python
# celula de carregamento do dataset e dataframe
df = pd.read_csv('/content/StudentPerformanceFactors.csv') # diretorio do arquivo csv (le o arquivo no parenteses com o pd), (o arquivo precisa ser uma string n esqueça das aspas)
df.head() # mostra as primeiras 5 linhas do nosso df (dataframe)
```

```python
# atributos numericos (quantitativos)
atributos_numericos = ['Hours_Studied',
                       'Attendance',
                       'Sleep_Hours',
                       'Previous_Scores',
                       'Tutoring_Sessions',
                       'Physical_Activity',
                       'Exam_Score']
```

```python
# variaveis categoricas binarias
variaveis_cat_binarias = ['Extracurricular_Activities',
                          'Internet_Access',
                          'School_Type',
                          'Learning_Disabilities',
                          'Gender']
```

```python
# variavies categoricas ordinais
variaveis_cat_ordinais = [ 'Parental_Involvement',
                          'Access_to_Resources',
                           'Motivation_Level',
                           'Family_Income',
                           'Teacher_Quality',
                           'Peer_Influence',
                           'Distance_from_Home']
```

```python
# rotulo do dataset
rotulo = 'Exam_Score'
```

```python
# indicadores estatísticos dos atributos numéricos
df[atributos_numericos + [rotulo]].describe()
```

```python
# listas de niveis de motivação
motivacao = df['Motivation_Level'].unique().tolist()
```

```python
# relação entre niveis de motivação e notas
for Motivation_Level in motivacao:
  print('Indicadores numéricos do ' + Motivation_Level + '\n')
  print(df[[rotulo]][(df['Motivation_Level'] == Motivation_Level)].describe(),'\n\n\n')
print('\n\n\n')
```

```python
# relação entre niveis de motivação e horas estudadas
for Motivation_Level in motivacao:
  print('Indicadores numéricos do ' + Motivation_Level + '\n')
  print(df[['Hours_Studied']][(df['Motivation_Level'] == Motivation_Level)].describe(),'\n\n\n')
print('\n\n\n')
```

```python
# relação entre niveis de motivação e presença
for Motivation_Level in motivacao:
  print('Indicadores numéricos do ' + Motivation_Level + '\n')
  print(df[['Attendance']][(df['Motivation_Level'] == Motivation_Level)].describe(),'\n\n\n')
print('\n\n\n')
```

```python
for variaveis_cat_binarias in variaveis_cat_binarias:
  plt.figure(figsize=(4, 3))

  sns.boxplot(x=variaveis_cat_binarias, y=rotulo, data=df[[variaveis_cat_binarias, rotulo]])
  plt.title('Distribuição por ' + variaveis_cat_binarias)
  plt.show()
```

```python
for variaveis_cat_ordinais in variaveis_cat_ordinais:
  plt.figure(figsize=(4, 3))

  sns.boxplot(x=variaveis_cat_ordinais, y=rotulo, data=df[[variaveis_cat_ordinais, rotulo]])
  plt.title('Distribuição por ' + variaveis_cat_ordinais)
  plt.show()
```

```python
# Correlação entre todos os atributos numéricos
# Hipótese: O atributo filhos é irrelevante para a previsão, visto que possui baixa correlação com o rótulo
plt.figure(figsize=(4, 3))
matriz_corr = df[atributos_numericos + [rotulo]].corr()
sns.heatmap(      # cria gráfico do tipo heatmap
matriz_corr,      # dados da matriz de correlação
annot=True,       # os valores são exibidos dentro de cada célula
cmap='coolwarm',  # cores do gráfico
fmt='.2f'         # quantidade de casas decimais
)
plt.title('Correlação entre Atributos númericos e rótulo', fontweight='bold')
plt.show()
```
